#**GffutilsAI**

**Generation of hypotheses regarding the function and evolution of vacuolar proteins**

The idea is to use LLM questions to transform the task of writing complex programmatic parsing scripts (in Python, Perl, or R) into a series of natural language database queries. This approach will make the GFF data more accessible and significantly speed up the future generation of hypotheses regarding the function and evolution of vacuolar proteins.

**Queries**

1.   I want a list of all attributes in GCA_050947715.1_ASM5094771v1_genomic.gff
2.   Tell me which species it is
3.   What is the geographical origin of this genome?
4.   Give me the number of chromosomes
5.   Give me the number of genes on each chromosome
6.   List all vacuolar proteins and their protein_id

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define the data
# This data includes the model name, its score with the agent, and its category.
data = [
    {'Model': 'Gemini 2.5 Pro', 'Score': 6, 'Type': 'Frontier (Large)'},
    {'Model': 'GLM 4.6', 'Score': 6, 'Type': 'Frontier (Large)'},
    {'Model': 'gpt-oss:120b', 'Score': 6, 'Type': 'Open Weights (Large)'},
    {'Model': 'Claude Opus 4.1', 'Score': 6, 'Type': 'Frontier (Large)'},
    {'Model': 'Claude 3.5 Haiku', 'Score': 6, 'Type': 'Efficient (Small)'},
    {'Model': 'Gemini 2.5 Flash', 'Score': 5, 'Type': 'Efficient (Small)'},
    {'Model': 'gpt-oss:20b-local', 'Score': 5, 'Type': 'Open Weights (Small)'},
    {'Model': 'Claude Sonnet 4.5', 'Score': 5, 'Type': 'Frontier (Mid)'},
    {'Model': 'gpt-oss:20b-cloud', 'Score': 4, 'Type': 'Open Weights (Small)'},
]

# 2. Create DataFrame
df = pd.DataFrame(data)

# 3. Sort the data
# Sort by Score (descending) and then by Model name for a clean layout.
df_sorted = df.sort_values(by=['Score', 'Model'], ascending=[False, True])

# 4. Define the color palette for the categories
colors = {
    'Frontier (Large)': '#2ecc71',  # Emerald Green
    'Open Weights (Large)': '#27ae60', # Nephritis Green
    'Frontier (Mid)': '#f1c40f',      # Sunflower Yellow
    'Efficient (Small)': '#3498db',   # Peter River Blue
    'Open Weights (Small)': '#2980b9' # Belize Hole Blue
}

# 5. Create the Plot
plt.figure(figsize=(12, 7))

# Create the stratified bar chart
barplot = sns.barplot(
    x='Score',
    y='Model',
    data=df_sorted,
    hue='Type',      # Color bars based on their 'Type'
    dodge=False,     # Keep bars centered on their y-tick
    palette=colors   # Use our custom color map
)

# 6. Add Labels, Title, and Legend
plt.title('Impact of Underlying LLM on Agent Performance', fontsize=16)
plt.xlabel('Agent Success Rate (out of 6)', fontsize=12)
plt.ylabel('Underlying Model', fontsize=12)

# Customize the legend
plt.legend(title='LLM Class', title_fontsize='12', loc='lower right', frameon=True)

# 7. Add numerical labels to the end of each bar
for i, p in enumerate(barplot.patches):
    width = p.get_width()
    # The 'dodge=False' creates some empty patches for spacing, skip them.
    if width > 0:
        plt.text(
            width + 0.1,       # x-position slightly after the bar end
            p.get_y() + p.get_height() / 2, # y-position centered on the bar
            f'{int(width)}/6', # The label text
            ha='left',         # Horizontal alignment
            va='center',       # Vertical alignment
            fontsize=10
        )

# 8. Final Touches
plt.xlim(0, 7) # Extend x-axis slightly for labels
plt.tight_layout()

# 9. Save and Show
plt.savefig('llm_impact.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

# 1. Define the Data
# 0 = Fail (Red), 1 = Success (Green)
# We pair them: Base (Standalone) vs Agent (With your code)
data_pairs = [
    # --- GLM 4.6 ---
    {'Model': 'GLM 4.6 [Base]',  'Q1':0, 'Q2':0, 'Q3':0, 'Q4':0, 'Q5':0, 'Q6':0},
    {'Model': 'GLM 4.6 [Agent]', 'Q1':1, 'Q2':1, 'Q3':1, 'Q4':1, 'Q5':1, 'Q6':1},

    # --- gpt-oss:120b ---
    {'Model': 'gpt-oss:120b [Base]',  'Q1':0, 'Q2':0, 'Q3':0, 'Q4':0, 'Q5':0, 'Q6':0},
    {'Model': 'gpt-oss:120b [Agent]', 'Q1':1, 'Q2':1, 'Q3':1, 'Q4':1, 'Q5':1, 'Q6':1},

    # --- gpt-oss:20b ---
    {'Model': 'gpt-oss:20b [Base]',  'Q1':0, 'Q2':0, 'Q3':0, 'Q4':0, 'Q5':0, 'Q6':0},
    {'Model': 'gpt-oss:20b [Agent]', 'Q1':1, 'Q2':1, 'Q3':1, 'Q4':1, 'Q5':1, 'Q6':0},

    # --- Haiku 4.5 (Claude 3.5 Haiku) ---
    {'Model': 'Haiku 4.5 [Base]',  'Q1':0, 'Q2':0, 'Q3':1, 'Q4':0, 'Q5':0, 'Q6':0},
    {'Model': 'Haiku 4.5 [Agent]', 'Q1':1, 'Q2':1, 'Q3':1, 'Q4':1, 'Q5':1, 'Q6':1},

    # --- Gemini 2.5 Flash ---
    {'Model': 'Gemini 2.5 Flash [Base]',  'Q1':0, 'Q2':0, 'Q3':1, 'Q4':0, 'Q5':0, 'Q6':0},
    {'Model': 'Gemini 2.5 Flash [Agent]', 'Q1':1, 'Q2':1, 'Q3':0, 'Q4':1, 'Q5':1, 'Q6':1},

    # --- Gemini 2.5 Pro ---
    {'Model': 'Gemini 2.5 Pro [Base]',  'Q1':0, 'Q2':1, 'Q3':1, 'Q4':1, 'Q5':0, 'Q6':0},
    {'Model': 'Gemini 2.5 Pro [Agent]', 'Q1':1, 'Q2':1, 'Q3':1, 'Q4':1, 'Q5':1, 'Q6':1},
]

# 2. Create DataFrame and Set Index
df = pd.DataFrame(data_pairs)
df = df.set_index('Model')

# 3. Define Custom Colors (Red/Green)
# #e74c3c = Alizarin Red (Fail)
# #2ecc71 = Emerald Green (Success)
cmap = ListedColormap(['#e74c3c', '#2ecc71'])

# 4. Create the Heatmap
plt.figure(figsize=(10, 8))

# Draw heatmap
# cbar=False turns off the color bar (we will make a custom legend)
# linewidths=1 adds the white grid lines
ax = sns.heatmap(df, cmap=cmap, annot=True, cbar=False,
                 linewidths=1, linecolor='white', square=False)

# 5. Add Titles and Labels
plt.title('The "Resurrection" Effect\nBaseline vs. Agentic Performance', fontsize=16, pad=20)
plt.xlabel('Question ID', fontsize=12)
plt.ylabel('Model Configuration', fontsize=12)

# 6. Create a Custom Legend
legend_elements = [
    Patch(facecolor='#e74c3c', edgecolor='white', label='Fail / Wrong (0)'),
    Patch(facecolor='#2ecc71', edgecolor='white', label='Success(1)')
]
plt.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, -0.08),
           ncol=2, frameon=False, fontsize=11)

# 7. Final Layout Adjustments
plt.tight_layout()

# 8. Save and Show
plt.savefig('heatmap_res.png', dpi=300)
plt.show()
```
